```python
import pandas as pd
import numpy as np
```

```python
data_path = "/Users/zhangchunyu/PycharmProjects/audio_etl/data/20181218"
#read_data(data_path)
data_source = pd.read_json(data_path,lines=True)
```

```python
print(data_source.shape,data_source.columns)
data_source.describe()
```

```python
effect_data = data_source[(data_source.teacher_speak_times>0) & (data_source.lesson_serial_number.str.find("MC")>=0)]
print(effect_data.shape)
effect_data.head()
```

```python
key = "teacher_speak_times"
pt = pd.pivot_table(effect_data,index=["lesson_serial_number"],
values=[key],#"communication_times","student_long_speak_length","student_long_speak_times",
        #"student_speak_length","student_speak_times","teacher_long_speak_length","teacher_long_speak_times",
       #"teacher_speak_length","teacher_speak_times"],
aggfunc=[np.mean,np.max,np.min,np.std,len])
#pt.columns
print("max exception")
print(pt[(pt[("len",key)]>30) & (pt[("amax",key)]> (pt[("mean",key)] + 3 * pt[("std",key)]))])
print("min exception")
print(pt[(pt[("len",key)]>30) & (pt[("amin",key)] < (pt[("mean",key)] + 3 * pt[("std",key)]))])
#pt
#print(pt.len.shape)
```

```python
key = "communication_times"
pt = pd.pivot_table(effect_data,index=["parent_feedback_class_level","parent_feedback_unit_name"],
values=["communication_times","student_speak_length","teacher_speak_length"],#"communication_times","student_long_speak_length","student_long_speak_times",
        #"student_speak_length","student_speak_times","teacher_long_speak_length","teacher_long_speak_times",
       #"teacher_speak_length","teacher_speak_times"],
aggfunc=[np.mean,np.max,np.min,np.std,len])
#print("max exception")
#print(pt[(pt[("len",key)]>30) & (pt[("amax",key)]> (pt[("mean",key)] + 3 * pt[("std",key)]))])
print("min exception")
print(pt[(pt[("len",key)]>30) 
         & (pt[("amin","communication_times")] < (pt[("mean","communication_times")] - 3 * pt[("std","communication_times")]))
         & (pt[("amin","student_speak_length")] < (pt[("mean","student_speak_length")] - 2 * pt[("std","student_speak_length")]))
         & (pt[("amin","teacher_speak_length")] < (pt[("mean","teacher_speak_length")] - 2 * pt[("std","teacher_speak_length")]))
        ])
#print(pt[(pt[("len",key)]>30) & (pt[("amin",key)] < (pt[("mean",key)] + 3 * pt[("std",key)]))])
```
